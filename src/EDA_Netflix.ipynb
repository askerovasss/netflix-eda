```python
Q1 = rated['user rating score'].quantile(0.25)
Q3 = rated['user rating score'].quantile(0.75)

IQR = Q3 - Q1

l_bounder = max(0, Q1 - 1.5 * IQR)
u_bounder = min(100, Q3 + 1.5 * IQR)

print(f'Q1 = {Q1:.2f}, Q3 = {Q3:.2f}, IQR = {IQR:.2f}')
print(f'Emissions limits: {l_bounder:.2f} and {u_bounder:.2f}')
```

##Hypothesis: “The influence of age rating (rating) on assessments (user rating score)”

H0: The average user rating score is the same for all rating levels (TV-14, TV-MA, PG, etc.).


H1: there are statistically significant differences.

```python
from scipy.stats import f_oneway #source: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.f_oneway.html, https://www.geeksforgeeks.org/how-to-perform-a-one-way-anova-in-python/

rated = data[data['has_rating'] == 1]
rating_groups = [rated.loc[rated['rating'] == lvl, 'user rating score'] for lvl in rated['rating'].unique()]

f_stat, p_value = f_oneway(*rating_groups)
print(f"ANOVA: F = {f_stat:.2f}, p = {p_value:.3f}")
```

Since p < 0.05, we reject the null hypothesis. Accordingly, age rating statistically influences the ratings given by users.

```python
#source: https://fastercapital.com/ru/content/HSD-Tukey--%D0%BF%D0%BE%D1%81%D0%BB%D0%B5-ANOVA.html
from statsmodels.stats.multicomp import pairwise_tukeyhsd

rated = data[data['has_rating']==1]

scores = rated['user rating score']
ages = rated['rating']

tukey = pairwise_tukeyhsd(endog=scores, groups=ages, alpha=0.05)
print(tukey)
```

How realistic is the result?
1. There is a correlation, but the effect is small.
F = 2.77 — moderate value: there is a difference between the groups, but it is not huge.
2. Post-hoc tests did not reveal any clear “leaders”. This suggests that the influence is spread across all categories rather than concentrated in one or two extreme groups.

The conclusion about the relationship can be trusted, but one should not expect any one rating label to “skyrocket” or “plummet.”

##Hypothesis: “The probability of no user rating score with the ratingLevel category”

```python
#source: https://success.openhealth.fr/ru/articles/4128369-%D1%82%D0%B5%D1%81%D1%82-%D0%BD%D0%B0-%D0%BD%D0%B5%D0%B7%D0%B0%D0%B2%D0%B8%D1%81%D0%B8%D0%BC%D0%BE%D1%81%D1%82%D1%8C

from scipy.stats import chi2_contingency

ct = pd.crosstab(data['ratingLevel'], data['has_rating'])
chi2, p, dof, expected = chi2_contingency(ct)

print("Chi2 =", round(chi2,2), " p-value =", round(p,4))
print("\nContingency (observed):\n", ct)
print("\nContingency (expected at H0):\n", pd.DataFrame(expected, index=ct.index, columns=ct.columns))
```

Since we obtained χ2 = 138.72 and p = 0.0035 (p < 0.05) in the χ2 test, we reject H0 about the independence of ratingLevel and the presence of user ratings (has_rating). That is, the probability that a show will receive a rating is statistically dependent on its age rating description.

Let's take a couple of examples from the “observed” and “expected” frequencies:

General Audiences (content for everyone) has significantly more exemptions (has_rating=0) — 34, whereas if it were independent, we would expect ~25.9. Conversely, fewer shows in this category receive ratings (19 vs. 27.1 expected). Mature Audiences (content for adults) is pretty much what we'd expect — the difference is small: 40 vs 39 and 40 vs 41.

#### Compare the Average User Rating in 2016 with the Average Rating in 2017
- Can it be concluded that 2017 was more successful for Netflix? ("More successful" means users gave higher ratings on average)
- How else can "success" be evaluated?



```python
from scipy.stats import ttest_ind
ratings_2016 = data[data['release year'] == 2016]['user rating score'].dropna()
ratings_2017 = data[data['release year'] == 2017]['user rating score'].dropna()

mean_2016 = ratings_2016.mean()
mean_2017 = ratings_2017.mean()

t_stat, p_val = ttest_ind(ratings_2016, ratings_2017, equal_var=False)

print(f"Average User Rating in 2016: {mean_2016:.2f}")
print(f"Average Rating in 2017: {mean_2017:.2f}")
print(f"T-stat: {t_stat:.2f}")
print(f"P-value: {p_val:.4f}")
```

Although the average score is higher in 2017, a statistical test (t-test) shows that the difference is not statistically significant (p > 0.05). “Success” can still be measured by the share of shows with high ratings, the number of unique popular genres, the level of engagement, ratings by age group, and others.

```python
threshold = 85 #I set a threshold for what qualifies as a “highly rated show.”
data['is_high_rated'] = data['user rating score'] > threshold

highrated_stats = data[data['release year'].isin([2016, 2017])]
highrated_by_year = (
    highrated_stats.groupby('release year')['is_high_rated']
    .agg(['mean', 'count'])
    .rename(columns={'mean': 'share_of_highrated_shows', 'count': 'total_shows'})
    .round(2)
)
print(highrated_by_year)
```

While the average score was higher in 2017, the proportion of actual highly rated shows is almost the same as in 2016. This once again confirms that the difference is not statistically significant.

```python
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer
text_data = data['ratingLevel'].fillna("").astype(str)

vectorizer = CountVectorizer(stop_words='english', max_df=0.95, min_df=2)
doc_term_matrix = vectorizer.fit_transform(text_data)

n_topics = 6
lda = LatentDirichletAllocation(n_components=n_topics, random_state=42)
lda.fit(doc_term_matrix)

def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print(f"\nTopic {topic_idx + 1}:")
        top_words = [feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]]
        print(", ".join(top_words))

print("LDA Topiks:")
display_topics(lda, vectorizer.get_feature_names_out(), 8)

topic_distribution = lda.transform(doc_term_matrix)
data['topic'] = topic_distribution.argmax(axis=1)

rating_by_topic = data.groupby('topic')['user rating score'].agg(['mean', 'count']).round(2)
rating_by_topic.columns = ['Average_rating', 'Number_of_shows']

print("Average User Rating by Topic:")
print(rating_by_topic)
```

Likely to be quality family or educational content. Topic 3 Relatively high ratings despite limitations. These topics are in demand with teen and adult audiences. Topic 4 This is the most popular topic by rating - versatility, combination of action and accessible Topic 5 Audience appreciates such shows - they are perceived as quality, in-depth, more “adult”. Topic 6 Audiences like the combination of action and humor, despite the “crudeness” of the themes. Topics 4, 5 and 6 receive the highest ratings: this is mature, universal and genre interesting content.

The grade for the group project as a whole — the first of the two grades — is given on a 10-point scale based on 5 criteria. Additionally, the committee may, at its discretion, add (but not decrease) a certain number of bonus points beyond these criteria.

Each of the criteria will be listed and reviewed below.

### Criterion 1: Technical Informativeness – It is necessary to identify non-trivial dependencies and trends in the data (up to 2 points)

This criterion evaluates how technically advanced you have become in identifying patterns, trends, and dependencies in the data. The more complex tools you apply during the project and the more complex and non-trivial dependencies you uncover, the higher your team's score will be for this criterion!

### Criterion 2: Analytical Informativeness – It is necessary to interpret the data and draw conclusions useful for the business (up to 2 points)

This criterion evaluates the practical significance of your actions, the interpretation of technical tools, the ability to highlight the usefulness and importance of the conclusions from a business perspective, and the ability to align the results with the original business goals and "sell" what you have accomplished.


### Criterion 3: Visualization – The technical execution of the charts is evaluated (up to 2 points)

This criterion assesses the complexity and technicality of your visualizations. Visualizations should be clear, aesthetically pleasing, understandable, easily interpretable, and non-trivial in terms of technical execution and the conclusions they convey.

*The maximum score for this criterion (**2 points**) cannot be achieved if only basic visualization libraries such as Matplotlib and Seaborn are used.*

### Criterion 4: Dataset Preprocessing – Data Cleaning/Replacement (up to 1 point)

This criterion evaluates the completeness of all actions performed in terms of data preparation and initial dataset preprocessing.


### Criterion 5: Data Enrichment – Adding New Features/Datasets (up to 3 points)

This criterion evaluates the extent to which feature engineering is used based on existing data, as well as data enrichment by supplementing the existing dataset with meaningful information from related external datasets, various external resources, and other sources. Note that related external information should not replace or substitute the original task and data in the presented dataset, but should only complement and enrich it. The correctness and depth of interaction with external resources are evaluated within this criterion!

*The maximum score for this criterion (**3 points**) can only be achieved with correct and meaningful work involving at least two sources of external data (two external datasets), as well as having at least two examples of valid and meaningful application of feature engineering.*

*For this criterion, **2 points** can only be achieved with correct and meaningful work involving at least one source of external data (an external dataset), as well as having at least one example of valid and meaningful application of feature engineering.*
